```python
#convert Numeric to Categorical
# linspace is used to determine the evenly spaced sequence in a specified interval
import numpy as np
import pandas as pd
path = "D:/TUTEDUDE/DateAnalysis/data.xlsx"
df1=pd.read_excel(path)
print(df1)
bins=np.linspace(min(df1["Length"]), max(df1["Length"]),4)
```

```python
group_names = ["Low", "Medium", "High"]
```

```python
df1["Length-binned"]=pd.cut(df1["Length"], bins, labels = group_names, include_lowest=True)
```

```python
df1.head(10)
```
